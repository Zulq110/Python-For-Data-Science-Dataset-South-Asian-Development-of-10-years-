```python
import pandas as pd
```

```python
file_path = 'South_Asian_dataset.csv'
data = pd.read_csv(file_path)
```

```python
selected_columns = data[["Country", "Year", "GDP (current US$)"]]
print(selected_columns.head())
```

```python
data["GDP per capita normalized"] = data["GDP per capita (current US$)"] / data["GDP per capita (current US$)"].max()
data
```

```python
data.replace("..", pd.NA, inplace=True)
data.fillna(data.mean(numeric_only=True), inplace=True)
data
```

```python
mean_gdp_per_year = data.groupby("Year")["GDP (current US$)"].mean()
print(mean_gdp_per_year)
```

```python
sorted_data = data.sort_values(by="GDP (current US$)", ascending=False)
print(sorted_data.head())
```

```python
import numpy as np
```

```python
gdp_array = np.array(data["GDP (current US$)"].dropna())
data
```

```python
gdp_normalized = gdp_array / gdp_array.max()
data
```

```python
top_5_gdp = gdp_array[:5]
middle_values = gdp_array[len(gdp_array)//2 - 2: len(gdp_array)//2 + 3]
data
```

```python
random_array = np.random.randint(1, 100, (3, 3))
random_array_sum = random_array.sum()
random_array_mean = random_array.mean()
data
```

```python
print("Top 5 GDPs:", top_5_gdp)
print("Middle values:", middle_values)
print("Random Array:\n", random_array)
print("Sum:", random_array_sum, "Mean:", random_array_mean)
```

```python
import matplotlib.pyplot as plt
```

```python
afghanistan_data = data[data["Country"] == "Afghanistan"]
plt.figure(figsize=(10, 6))
plt.plot(afghanistan_data["Year"], afghanistan_data["GDP (current US$)"], marker='o', label='Afghanistan GDP')
plt.title("GDP Over Years for Afghanistan")
plt.xlabel("Year")
plt.ylabel("GDP (current US$)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
top_5_gdp_countries = sorted_data[["Country", "GDP (current US$)"]].head(5)
plt.figure(figsize=(10, 6))
plt.bar(top_5_gdp_countries["Country"], top_5_gdp_countries["GDP (current US$)"], color='green')
plt.title("Top 5 Countries by GDP")
plt.xlabel("Country")
plt.ylabel("GDP (current US$)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(gdp_array / 1e12, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title("GDP Distribution (in Trillions)")
plt.xlabel("GDP (Trillions of US$)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data["GDP (current US$)"], data["GDP growth (annual %)"], alpha=0.5, color='red')
plt.title("GDP Growth vs GDP")
plt.xlabel("GDP (current US$)")
plt.ylabel("GDP Growth (annual %)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(top_5_gdp_countries["GDP (current US$)"], labels=top_5_gdp_countries["Country"], autopct='%1.1f%%',
        startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink', 'orange'])
plt.title("GDP Proportion of Top 5 Countries")
plt.show()
```
